### Join driver data
- filter on hours over 100
- display toplist as a barchart: while Turi Create has it's own data visualization tools, it also can convert to Pandas dataframes, so we display both charts.

```python
lap_times_sum_filtered_sf = lap_times_sum_sf[lap_times_sum_sf['hours'] >= 100]
```

```python
lap_times_joined_sf = lap_times_sum_filtered_sf.join(drivers_sf, on={'driverId':'driverId'}, how='inner').sort('hours', ascending = False)
lap_times_joined_sf.head(5)
```

```python
turicreate.show(lap_times_joined_sf['driverRef'], lap_times_joined_sf['hours'], xlabel='driverRef', ylabel='hours', title=None)
```

```python
lap_times_joined_sf.to_dataframe().plot.barh(y='hours', x='driverRef', figsize=(14,10))
```
